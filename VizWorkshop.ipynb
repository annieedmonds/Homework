```python
# Python code using Seaborn and the Social Network Edges Data Set
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Load the data set
social_edges = pd.read_csv('social_network_edges.csv')

sns.pairplot(social_edges, hue='target')
plt.title('Seaborn Pairplot of Social Network Dataset')
plt.show()

```

```python
# Python code using Matplotlib and the Airbnb data set
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('airbnb_dataset.csv')

# Generate sample data
x = df['Price']
y = df['Rating']

# Create a basic scatter plot
plt.scatter(x, y, color = 'blue', s=100)
plt.title("Price and Rating of Airbnbs Scatter Plot")
plt.xlabel("Price")
plt.ylabel("Rating")

```
