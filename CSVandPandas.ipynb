```python
#import pandas
import pandas as pd
df = pd.read_csv('/content/acc_players-2324F.csv', header=1)
df.head(5)
```

```python
#basic information about the datasets
print("Dataset Information:")
print(df.info())
```

```python
#Check for missing values
print("\nMissing Values:")
df.isnull().sum()
```

```python
#drop rows with missing values
df = df.dropna()
print("\nMissing Values:")
df.isnull().sum()

#Rename columns
df.rename(columns = {'TRB':'Rebounds'}, inplace = True)
```

```python
#Basic Analysis:
#-Calculate the total points by all players combined
total_points = df['PTS'].sum()
print("\nTotal Points scored:", total_points)
#-Display the player who has played the most minutes (MP)
most_minutes = df.loc[df['MP'].idxmax(), 'Player']
print("\nPlayer with most minutes played:", most_minutes)

#-Identify the top 5 players in terms of total rebounds (TRB)
top_rebounds = df.sort_values(by='TRB', ascending = False).head(5)
print("\nTop 5 players with the most rebounds:")
top_rebounds
```

```python
#Player Filtering
#-Create a new DataFrame containing only players who played more than 500 minutes
players_500 = df[df['MP']>500]
print("\nPlayers who played more than 500 minutes:")
players_500
```

```python

#-From this filtered DataFrame, determine the player with the highest total assists (AST)
most_assists = players_500.loc[players_500['AST'].idxmax(), 'Player']
print("\n Player with the highest total assists:")
most_assists
```

```python
#-Who are the top 3 Assist Leads in the League?
top3_assists = df.sort_values(by = 'AST', ascending = False).head(3)
print("\nTop 3 Assist Leads in the League:")
top3_assists
```

```python
#-Who are the top 3 shot blockers?
top_shotblockers = df.sort_values(by = 'BLK', ascending = False).head(3)
print("\nTop 3 Shot Blockers:")
top_shotblockers
```

```python
#School Based Analysis
#-Group the players by school and calculate the total points scored by each school
total_points_bySchool = df.groupby('School')[['PTS']].sum().reset_index()
print("\nTotal Points Scored by each School")
total_points_bySchool
```

```python
#-Group the players by School and Calculate the total assists (AST) for each team
total_assists_bySchool = df.groupby('School')[['AST']].sum().reset_index()
print("\nTotal Assists per Team:")
total_assists_bySchool
```

```python
#-Sort the schools by total points scored and display the top 3 schools
top3_highestscoring = total_points_bySchool.sort_values(by = 'PTS', ascending = False).head(3)
print("\nTop 3 Schools:")
top3_highestscoring
```
